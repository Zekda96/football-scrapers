```python
import soccerdata as sd
from clean_dataframe import clean_df

leagues = 'ESP-La Liga'
seasons = '23-24'
ws = sd.WhoScored(leagues=leagues, seasons=seasons)
events = ws.read_events(match_id=1734879, force_cache=True)
events
```

```python
qualifiers = ["own_goal", "angle", "length", "throw_in"
              , "big_chance"
              , "big_chance_created", "key_pass"
              , "head", "long_ball"
              , "goal_kick", "free_kick", "freekick_taken", "corner_taken", "cross"
              , "blocked_cross"
              , "zone"
              , "right_foot", "left_foot"
              , "yellow_card", "second_yellow", "red_card"
              ,"leading_to_goal"
              ]
df = clean_df(events, qualifiers)
df
```

```python
import pandas as pd

h = df["home"].unique()[0]
a = df["away"].unique()[0]
fp = f"data/{leagues}_{seasons}/{h}_{a}.json"

# DataFrame to json table
result = df.to_json(fp, orient="table")

# Read json table as DataFrame
xdf = pd.read_json(fp, orient="table")

xdf
```
